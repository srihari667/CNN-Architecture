## CNN Architecture

1. What is the role of filters and feature maps in Convolutional Neural Network (CNN)?

    In a **Convolutional Neural Network (CNN)**, **filters** and **feature maps** play crucial roles in understanding and extracting patterns from images (or other spatial data). Here’s a clear explanation:


    ##  **1. Filters (Kernels)**

    A **filter** is a small matrix of weights (e.g., 3×3, 5×5) that slides over the input image.

    ### **Role of Filters**

    * **Extract features** from the input such as:

      * edges
      * corners
      * textures
      * shapes
    * Filters learn to detect different patterns during training.
    * Each filter focuses on **one specific type of feature**.

      * For example, one filter may learn vertical edges, another horizontal edges.

    ### **How it works**

    * The filter performs **convolution**: multiplies its values with the local region of the input image and sums up the result.
    * These operations help detect **spatial relationships** between pixels.


    ##  **2. Feature Maps**

    A **feature map** (also called activation map or output map) is the result obtained after applying a filter to the input.

    ### **Role of Feature Maps**

    * They **represent the presence and location of the learned features** in the input.
    * Each feature map highlights where the filter detected its pattern.
    * The deeper layers produce more complex feature maps (e.g., shapes, objects).

2. Explain the concepts of padding and stride in CNNs(Convolutional Neural Network). How do they affect the output dimensions of feature maps?



    #  **1. Padding in CNN**

    **Padding** means adding extra rows/columns (usually zeros) around the border of the input image.

    ###  **Why Padding is Used?**

    1. **Preserve spatial dimensions**

       * Without padding, convolution reduces the size of the image.
       * Padding helps keep output size the same as input.

    2. **Better feature extraction at borders**

       * Without padding, edge pixels get fewer operations.

    3. **Control output dimensions**

       * More padding → larger output feature map.


    ### **Types of Padding**

    * **Valid Padding (No padding)**

      * Output becomes smaller.
    * **Same Padding (Zero padding)**

      * Output size ≈ input size.


    ###  **Effect of Padding on Output Size**

    Formula for output height/width:

    [
    \text{Output} = \frac{(N - F + 2P)}{S} + 1
    ]

    Where:

    * (N) = input size
    * (F) = filter size
    * (P) = padding
    * (S) = stride


    #  **2. Stride in CNN**

    **Stride** is how many pixels the filter moves (slides) at each step.

    ###  **Why Stride is Important?**

    * Controls how “fast” the filter moves across the image.
    * Larger stride → smaller output feature map (downsampling).
    * Smaller stride → larger output feature map (more detailed).


    ###  **Effect of Stride on Output Size**

    * **Stride = 1** → maximum overlap → larger output
    * **Stride = 2** → filter jumps → output becomes smaller

    Using the same formula:

    [
    \text{Output} = \frac{(N - F + 2P)}{S} + 1
    ]


    #  **Example: Understanding Effects**

    ### Suppose:

    * Input = 32 × 32
    * Filter = 3 × 3
    * Padding = 1
    * Stride = 1

    [
    \text{Output} = \frac{(32 - 3 + 2×1)}{1} + 1 = 32
    ]

     Output remains **32 × 32** (same padding).



3. Define receptive field in the context of CNNs. Why is it important for deep architectures?



    # **Receptive Field in CNNs**

    In a Convolutional Neural Network (CNN), the **receptive field** refers to the **region of the input image** that a particular neuron in any layer can “see” or is influenced by.
    It represents how much of the original input contributes to the activation of that neuron.


    # **Why It Is Important for Deep Architectures**

    1. **Captures Larger Context**
       As the network becomes deeper, the receptive field increases.
       This allows deeper layers to capture **global features** rather than local ones, improving the network’s understanding of the overall image structure.

    2. **Enables High-Level Feature Learning**

       * Early layers learn edges and textures (small receptive field).
       * Deep layers learn shapes, objects, and semantics (large receptive field).
         A large receptive field is necessary for recognizing complex patterns.

    3. **Improves Performance on Vision Tasks**
       Tasks like object detection, segmentation, and classification need information from **wide regions** of the image.
       A large receptive field helps the model make more accurate predictions.

    4. **Supports Spatial Relationship Understanding**
       Larger receptive fields allow the network to understand how different parts of an image relate to each other, which is crucial for detecting whole objects.


4. Discuss how filter size and stride influence the number of parameters in a CNN.


    ## **1. Filter Size and Number of Parameters**

    The **filter size directly affects the number of learnable parameters** in a CNN.

    ### **Why?**

    Each filter has:
    [
    \text{Parameters} = (\text{filter height}) \times (\text{filter width}) \times (\text{input channels}) + 1\ (\text{bias})
    ]

    ### **Larger filters → more parameters**

    Example:

    * 3×3 filter with 3 channels:
      [
      3 \times 3 \times 3 = 27 \text{ weights}
      ]
    * 5×5 filter with 3 channels:
      [
      5 \times 5 \times 3 = 75 \text{ weights}
      ]


    
      **Increasing filter size increases parameters**

      **Decreasing filter size reduces parameters**


    ## **2. Stride and Number of Parameters**

    **Stride does NOT affect the number of parameters.**

    ### Why?

    * Stride only controls **how the filter moves** over the image.
    * It does **not change** the filter weights or their count.

    Whether stride = 1, 2, or 3, the filter size stays the same → same number of parameters.

    ### Stride affects:

    * Output feature map size
    * Computational cost

    But **not the number of learnable weights**.


5. Compare and contrast different CNN-based architectures like LeNet,AlexNet, and VGG in terms of depth, filter sizes, and performance.


    ## **1. LeNet **

    ### **Depth**

    * Very shallow
    * ~5–7 layers total (including convolution + pooling + fully connected layers)

    ### **Filter Sizes**

    * Conv layers use **5×5 filters**
    * No 3×3 filters
    * Simpler architecture designed for small images (28×28)

    ### **Performance**

    * Designed for digit recognition (MNIST)
    * Works well on small grayscale images
    * Very low computational cost
    * Not suitable for large-scale datasets like ImageNet


    ## **2. AlexNet (2012)**

    ### **Depth**

    * Much deeper than LeNet
    * **8 layers** (5 convolution + 3 fully connected)

    ### **Filter Sizes**

    * First layer uses **11×11 filters**
    * Later layers use **5×5 and 3×3 filters**
    * Introduced ReLU activation and dropout to improve performance

    ### **Performance**

    * First architecture to win ImageNet with a huge accuracy improvement
    * Showed the power of GPUs for training
    * Much better performance on large-scale image classification tasks
    * Reduced error rate significantly (TOP-5 error: ~15%)


    ## **3. VGG (2014)**

    ### **Depth**

    * Very deep
    * Versions: **VGG-11, VGG-16, VGG-19**
    * Up to **19 layers**

    ### **Filter Sizes**

    * Uses **only 3×3 filters** throughout the network
    * Uniform architecture makes it simple and modular
    * Stacked 3×3 filters increase receptive field (same as 5×5 or 7×7 but fewer parameters)

    ### **Performance**

    * Higher accuracy than AlexNet
    * Strong feature extractor used widely in transfer learning
    * Very computationally expensive (138M parameters in VGG16)
    * Performs extremely well on ImageNet


    # **Key Differences**

    ###  Depth

    * **LeNet** < **AlexNet** < **VGG (deepest)**

    ###  Filter Size

    * **LeNet:** large 5×5
    * **AlexNet:** mix (11×11 → 5×5 → 3×3)
    * **VGG:** only 3×3 small filters (stacked)

    ###  Performance

    * **LeNet:** simple tasks (digits)
    * **AlexNet:** breakthrough for large datasets
    * **VGG:** very high accuracy, used in transfer learning but heavy

```python
'''
7. Load and preprocess the CIFAR-10 dataset using Keras, and create a
CNN model to classify RGB images. Show your preprocessing and architecture.
'''


import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical
import numpy as np

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

x_train = x_train.astype('float32') / 255.0
x_test  = x_test.astype('float32')  / 255.0

num_classes = 10
y_train_cat = to_categorical(y_train, num_classes)
y_test_cat  = to_categorical(y_test, num_classes)


mean = np.mean(x_train, axis=(0,1,2), keepdims=True)
std  = np.std(x_train, axis=(0,1,2), keepdims=True)
x_train = (x_train - mean) / (std + 1e-7)
x_test  = (x_test  - mean) / (std + 1e-7)

data_augmentation = models.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.08),
    layers.RandomTranslation(0.06, 0.06),
    layers.RandomZoom(0.06)
], name="data_augmentation")

def build_cifar10_cnn(input_shape=(32,32,3), num_classes=10):
    inputs = layers.Input(shape=input_shape)
    x = data_augmentation(inputs)
    x = layers.Rescaling(1.0)(x)

    x = layers.Conv2D(32, (3,3), padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(32, (3,3), padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.Dropout(0.2)(x)

    x = layers.Conv2D(64, (3,3), padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(64, (3,3), padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Conv2D(128, (3,3), padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(128, (3,3), padding='same', activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.Dropout(0.4)(x)


    x = layers.Flatten()(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = models.Model(inputs, outputs, name="cifar10_cnn")
    return model

model = build_cifar10_cnn()
model.summary()

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

EPOCHS = 50
BATCH_SIZE = 128
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-6)
]

history = model.fit(
    x_train, y_train_cat,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=0.1,
    callbacks=callbacks,
    verbose=2
)

test_loss, test_acc = model.evaluate(x_test, y_test_cat, verbose=2)
print(f"Test accuracy: {test_acc:.4f}  |  Test loss: {test_loss:.4f}")
```

```python
'''
8.  Using PyTorch, write a script to define and train a CNN on the MNIST
dataset. Include model definition, data loaders, training loop, and accuracy evaluation.
'''


import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader
import os

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset  = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True, num_workers=2, pin_memory=True if device.type=='cuda' else False)
test_loader  = DataLoader(test_dataset, batch_size=256, shuffle=False, num_workers=2, pin_memory=True if device.type=='cuda' else False)

class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64*7*7, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 10)
        )
    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

model = SimpleCNN().to(device)
print(model)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

def train_one_epoch(model, loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * images.size(0)
        _, preds = outputs.max(1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
    return running_loss / total, correct / total

def evaluate(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            running_loss += loss.item() * images.size(0)
            _, preds = outputs.max(1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
    return running_loss / total, correct / total

EPOCHS = 5
for epoch in range(1, EPOCHS+1):
    train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, device)
    test_loss, test_acc = evaluate(model, test_loader, criterion, device)
    print(f"Epoch {epoch}/{EPOCHS} | Train Loss: {train_loss:.4f} Acc: {train_acc:.4f} | Test Loss: {test_loss:.4f} Acc: {test_acc:.4f}")

os.makedirs('/mnt/data', exist_ok=True)
torch.save(model.state_dict(), '/mnt/data/mnist_cnn_pytorch.pth')
print("Saved model to /mnt/data/mnist_cnn_pytorch.pth")

```

```python
'''
 Given a custom image dataset stored in a local directory, write code using
Keras ImageDataGenerator to preprocess and train a CNN model.
'''

import tensorflow as tf
import tensorflow_datasets as tfds
from pathlib import Path
from PIL import Image
import numpy as np
import random
import matplotlib.pyplot as plt
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models, callbacks


OUT_DIR = Path("/content/cats_and_dogs_tfds")
IMG_SIZE = (160,160)
BATCH_SIZE = 32
RANDOM_SEED = 42
NUM_TRAIN = 18000
NUM_VAL = 3000
NUM_TEST = 2000
USE_TRANSFER_LEARNING = True
EPOCHS = 8


random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

OUT_DIR.mkdir(parents=True, exist_ok=True)

print("Downloading cats_vs_dogs via tensorflow_datasets (this may be large; please be patient)...")
ds = tfds.load('cats_vs_dogs', split='train', shuffle_files=True, as_supervised=True)
info = tfds.builder('cats_vs_dogs').info
print("Total examples (tfds info):", info.splits['train'].num_examples)

def save_images_from_tfds(ds_iter, start_idx, end_idx, split_name):
    saved = 0
    for idx in range(start_idx, end_idx):
        try:
            img, label = next(ds_iter)
        except StopIteration:
            break
        cls = "cat" if int(label) == 0 else "dog"
        dest = OUT_DIR / split_name / cls
        dest.mkdir(parents=True, exist_ok=True)

        img_arr = tf.image.resize(img, IMG_SIZE).numpy().astype(np.uint8)
        im = Image.fromarray(img_arr)
        im.save(dest / f"{cls}_{idx}.jpg", quality=90)
        saved += 1
    print(f"Saved {saved} images to {split_name}.")


ds = ds.shuffle(30000, seed=RANDOM_SEED)
ds_iter = iter(tfds.as_numpy(ds))


save_images_from_tfds(ds_iter, 0, NUM_TRAIN, "train")
save_images_from_tfds(ds_iter, NUM_TRAIN, NUM_TRAIN + NUM_VAL, "val")
save_images_from_tfds(ds_iter, NUM_TRAIN + NUM_VAL, NUM_TRAIN + NUM_VAL + NUM_TEST, "test")

print("Directory structure created at:", OUT_DIR)
print("Train cats:", len(list((OUT_DIR/"train"/"cat").glob("*.jpg"))) ,
      "Train dogs:", len(list((OUT_DIR/"train"/"dog").glob("*.jpg"))))


train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.12,
    height_shift_range=0.12,
    shear_range=0.12,
    zoom_range=0.12,
    horizontal_flip=True,
    fill_mode='nearest'
)
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    OUT_DIR / "train",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    seed=RANDOM_SEED
)

validation_generator = val_datagen.flow_from_directory(
    OUT_DIR / "val",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)

test_generator = val_datagen.flow_from_directory(
    OUT_DIR / "test",
    target_size=IMG_SIZE,
    batch_size=1,
    class_mode='binary',
    shuffle=False
)

def build_simple_cnn(input_shape=IMG_SIZE+(3,)):
    model = models.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(32, (3,3), activation='relu'),
        layers.MaxPooling2D(2,2),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D(2,2),
        layers.Conv2D(128, (3,3), activation='relu'),
        layers.MaxPooling2D(2,2),
        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def build_mobilenetv2_transfer(input_shape=IMG_SIZE+(3,), fine_tune_at=100):
    base_model = tf.keras.applications.MobileNetV2(input_shape=input_shape, include_top=False, weights='imagenet')
    base_model.trainable = True
    for layer in base_model.layers[:fine_tune_at]:
        layer.trainable = False
    inputs = layers.Input(shape=input_shape)
    x = tf.keras.applications.mobilenet_v2.preprocess_input(inputs)
    x = base_model(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.4)(x)
    outputs = layers.Dense(1, activation='sigmoid')(x)
    model = tf.keras.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

if USE_TRANSFER_LEARNING:
    print("Using MobileNetV2 transfer learning model (recommended).")
    model = build_mobilenetv2_transfer()
else:
    print("Using simple CNN model.")
    model = build_simple_cnn()

model.summary()


early = callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator,
    callbacks=[early]
)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history.history.get('accuracy', []), label='train_acc')
plt.plot(history.history.get('val_accuracy', []), label='val_acc')
plt.title('Accuracy'); plt.xlabel('epoch'); plt.ylabel('accuracy'); plt.legend()
plt.subplot(1,2,2)
plt.plot(history.history.get('loss', []), label='train_loss')
plt.plot(history.history.get('val_loss', []), label='val_loss')
plt.title('Loss'); plt.xlabel('epoch'); plt.ylabel('loss'); plt.legend()
plt.show()

if test_generator.samples > 0:
    loss, acc = model.evaluate(test_generator, steps=test_generator.samples, verbose=1)
    print(f"Test loss: {loss:.4f}, Test accuracy: {acc:.4f}")
else:
    print("No test samples found. Skipping test evaluation.")

```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

```python
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    zoom_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    data_dir + "/train",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

val_generator = test_datagen.flow_from_directory(
    data_dir + "/val",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    data_dir + "/test",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)
```

```python
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224,224,3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False  # freeze layers

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10
)
```

```python
model.evaluate(test_generator)
```

```python
model.save("/content/drive/MyDrive/chest_xray_model.keras")
```

```python
%%writefile app.py
import streamlit as st
import tensorflow as tf
from PIL import Image
import numpy as np

# Load your saved Keras model
model = tf.keras.models.load_model("chest_xray_model.keras")

st.title("Chest X-ray Classification App")
st.write("Upload a chest X-ray image to classify it as Normal or Pneumonia.")

uploaded_file = st.file_uploader("Choose an image...", type=["jpg", "jpeg", "png"])

def preprocess_image(img):
    img = img.resize((224, 224))
    img = np.array(img) / 255.0
    return np.expand_dims(img, axis=0)

if uploaded_file is not None:
    image = Image.open(uploaded_file)
    st.image(image, caption="Uploaded X-ray", use_column_width=True)

    img = preprocess_image(image)
    prediction = model.predict(img)[0][0]

    if prediction > 0.5:
        st.error(f"Result: Pneumonia ({prediction:.2f})")
    else:
        st.success(f"Result: Normal ({1 - prediction:.2f})")
```
